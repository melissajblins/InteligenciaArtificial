## Questão 4 - Busca Expectimax

> Modelando o problema com a técnica Expectimax, responda qual deve ser a ação escolhida pelo Max para começar a luta.
 

1. As duas ações possíveis para cada um dos jogadores (Max e Min) é socar e bloquear. 
2. A pontuação do jogo é a “vida” do Max menos a “vida” do Min. 
3. Suponha que a vida de ambos inicia em 10 e que a luta dura apenas 2 golpes (1 de Max e 1 de Min, nesta ordem). 
4. Considere que a probabilidade do Min socar é 0.8 e bloquear, 0.2.
5. Um soco não bloqueado tira 10 pontos da “vida” do jogador
6. Um bloqueio só tem efeito se ele for realizado antes do soco.

### Trabalhando com as informações

*   Pontuação (Vida Max - Vida Min):
  * Início: 10 - 10 = 0
  * Se Max ganha: 10 - 0 = 10
  * Se Min ganha: 0 - 10 = -10
  * Se empata: 10 - 10 = 0

*  Probabilidade Min:
  * Socar: 0.8
  * Bloquear: 0.2

*  v Min:
   (0.8 * 1) + (0.2 * 0) = 0.8

Ao analisar os modelos acima, fica claro que a jogada inicial de Max deve ser socar, uma vez que ao bloquear o resultado final sempre será o empate, já que são somente duas jogadas para cada um. Por outro lado, ao socar, ele tem a chance de que o Min também soque, sendo essa probabilidade de 0.8, e assim poderá defender-se em uma próxima jogada (se tiver).
